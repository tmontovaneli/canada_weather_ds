```
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

ds = pd.read_csv("/content/canada_weather.csv", encoding = "utf-8")
ds = ds.set_index('Month')

length = 13

station = ['CALGARY INTL A', 'WINNIPEG INTL A', 'OTTAWA INTL A']
stations = ds.loc[ds['Stn_Name'].isin(station)]

plt.figure(figsize=(14, 7))
sns.set_style("whitegrid")

# Mean Temperature (°C)
ax = sns.lineplot(data=stations, x="Month", y="Tm", hue="Stn_Name")
ax.set_title('Mean Temperature (°C)')
ax.set_xticks(range(length)) 

ax.plot()
```

```
plt.figure(figsize=(14, 7))

# Highest Monthly Maximum Temperature (°C)
ax = sns.lineplot(data=stations, x="Month", y="Tx", hue="Stn_Name")
ax.set_title('Highest Monthly Maximum Temperature (°C)')
ax.set_xticks(range(length)) 
```

```
plt.figure(figsize=(14, 7))

# Lowest Monthly Minimum Temperature (°C)
ax = sns.lineplot(data=stations, x="Month", y="Tn", hue="Stn_Name")
ax.set_title('Lowest Monthly Minimum Temperature (°C)')
ax.set_xticks(range(length)) 
```
