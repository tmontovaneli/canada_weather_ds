```
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

data_file = "https://raw.githubusercontent.com/tmontovaneli/canada_weather_ds/main/data/merged/canada_weather.csv"
```

```
NR_GRAPH_COLUMNS = 13

def load_file():
  ds = pd.read_csv(data_file, 
                 encoding = "utf-8")
  ds = ds.set_index('Month')
  return ds

def query_by_year(ds, year):
  new_ds = ds.query("Year == @year")
  return new_ds

def get_stations():
  return ['CALGARY INTL A', 'WINNIPEG THE FORKS', 'OTTAWA INTL A']

def plot_mean_temperature(ds, year):
  plt.figure(figsize=(14, 7))
  sns.set_style("whitegrid")

  # Mean Temperature (°C)
  ax = sns.lineplot(data=ds.loc[ds['Stn_Name'].isin(get_stations())], x="Month", y="Tm", hue="Stn_Name")
  ax.set_title('Mean Temperature (°C) - {0}'.format(year))
  ax.set_xticks(range(NR_GRAPH_COLUMNS)) 

  ax.plot()

def plot_max_temperature(ds):
  plt.figure(figsize=(14, 7))

  # Highest Monthly Maximum Temperature (°C)
  ax = sns.lineplot(data=ds.loc[ds['Stn_Name'].isin(get_stations())], x="Month", y="Tx", hue="Stn_Name")
  ax.set_title('Highest Monthly Maximum Temperature (°C)')
  ax.set_xticks(range(NR_GRAPH_COLUMNS)) 

def plot_min_temperature():
  plt.figure(figsize=(14, 7))

  # Lowest Monthly Minimum Temperature (°C)
  ax = sns.lineplot(data=ds.loc[ds['Stn_Name'].isin(get_stations())], x="Month", y="Tn", hue="Stn_Name")
  ax.set_title('Lowest Monthly Minimum Temperature (°C)')
  ax.set_xticks(range(NR_GRAPH_COLUMNS)) 
```

```
# Mean
ds = load_file()

year = 2018

ds = query_by_year(ds, year)

# mean
plot_mean_temperature(ds, year)
```

```
# max temperature
plot_max_temperature(ds)
```

```
# min temperature
plot_min_temperature()
```
